```python
col_names=['CS','Density','WC','pH','EC','F','G','Pollen_Analysis','Viscosity','Purity','Price']
# CS = Color Score
# WC = Water Content
# EC = Electrical Conductivity
# F = Fructose Level
# G = Glucose Level
```

```python
data=pd.read_csv('honey_price_dataset.csv',header=None,names=col_names)
data.head()
```

```python
data_reg,data_clf=data,data
data_reg.head()
```

```python
data_clf.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, accuracy_score
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
from catboost import CatBoostClassifier
```

```python
le=LabelEncoder()
le.fit(data_reg['Pollen_Analysis'])
le.fit(data_clf['Pollen_Analysis'])
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
```

```python
le_name_mapping
```

```python
data_reg['Pollen_Analysis']=le.fit_transform(data_reg['Pollen_Analysis'])
data_reg.head()
```

```python
data_clf['Pollen_Analysis']=le.fit_transform(data_clf['Pollen_Analysis'])
data_clf.head()
```

```python
y_reg=data_reg['Purity']
X_reg=data_reg.drop(columns=['Purity'])
```

```python
y_clf=data_clf['Pollen_Analysis']
X_clf=data_clf.drop(columns=['Pollen_Analysis'])
```

```python
X_train_reg,X_test_reg,y_train_reg,y_test_reg= train_test_split(X_reg,y_reg,test_size=0.2,random_state=42)
X_train_clf,X_test_clf,y_train_clf,y_test_clf= train_test_split(X_clf,y_clf,test_size=0.2,random_state=42)
```

```python
clf_params = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}

clf = RandomForestClassifier(random_state=42)
clf_grid = GridSearchCV(clf, clf_params, cv=10, scoring='accuracy')
clf_grid.fit(X_train_clf, y_train_clf)

best_clf = clf_grid.best_estimator_

# Evaluate
y_class_pred = best_clf.predict(X_test_clf)
clf_accuracy = accuracy_score(y_test_clf, y_class_pred)
print("Best Classifier Accuracy:", clf_accuracy)
```

```python
xgb_params = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 1],
    'colsample_bytree': [0.7, 1],
}
```

```python
xgb_clf = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')

xgb_grid = GridSearchCV(xgb_clf, xgb_params, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
xgb_grid.fit(X_train_clf, y_train_clf)

best_xgb_clf = xgb_grid.best_estimator_

# Predict and evaluate
y_pred_xgb = best_xgb_clf.predict(X_test_clf)
xgb_accuracy = accuracy_score(y_test_clf, y_pred_xgb)
print("XGBoost Classifier Accuracy:", xgb_accuracy)
```

```python
cat_params = {
    'iterations': [100, 200],
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.1],
    'l2_leaf_reg': [1, 3, 5],
}
```

```python
cat_clf = CatBoostClassifier(verbose=0, random_state=42)

cat_grid = GridSearchCV(cat_clf, cat_params, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
cat_grid.fit(X_train_clf, y_train_clf)

best_cat_clf = cat_grid.best_estimator_

# Predict and Evaluate
y_pred_cat = best_cat_clf.predict(X_test_clf)
cat_accuracy = accuracy_score(y_test_clf, y_pred_cat)
print("CatBoost Classifier Accuracy:", cat_accuracy)
```

```python
import joblib
# Save model
# XGBoost Accuracy - 87.16%
# CatBoost Accuracy - 82.86%
# Random Forest Accuracy - 77.91%
joblib.dump(best_xgb_clf, 'xgb_pollen_classifier.pkl')
# loaded_model = joblib.load('xgb_pollen_classifier.pkl')
```

```python
from xgboost import XGBRegressor
```

```python
# Define parameter grid
param_grid = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [3, 5, 7, 10],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.25, 0.5],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [1, 1.5, 2]
}

xgb_reg = XGBRegressor(random_state=42)

# Setup Randomized Search with 5-fold cross-validation
random_search = RandomizedSearchCV(
    estimator=xgb_reg,
    param_distributions=param_grid,
    n_iter=50,
    scoring='r2',
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Fit the model
random_search.fit(X_train_reg, y_train_reg)

# Best parameters
print("Best Params:", random_search.best_params_)
```

```python
X_train_reg.columns
```

```python
from sklearn.metrics import mean_squared_error, r2_score

best_model = random_search.best_estimator_

y_pred = best_model.predict(X_test_reg)

# Evaluation
mse = mean_squared_error(y_test_reg, y_pred)
r2 = r2_score(y_test_reg, y_pred)

print(f"Final MSE: {mse:.2f}")
print(f"Final RÂ² Score: {r2:.2f}")
```

```python
joblib.dump(best_model, 'xgb_purity_predictor.pkl')
```

```python
data_price=data
data_price.head()
```

```python
y_price=data_price['Price']
data_price=data_price.drop(columns=['Price'])
X_price_train,X_price_test,y_price_train,y_price_test=train_test_split(data_price,y_price,test_size=0.2,random_state=42)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.3, 0.5],
    'reg_alpha': [0, 0.1, 0.5],
    'reg_lambda': [1, 1.5, 2.0]
}
```

```python
xgb_reg = XGBRegressor(random_state=42)

random_search = RandomizedSearchCV(
    estimator=xgb_reg,
    param_distributions=param_grid,
    n_iter=50,                # Try 50 random combinations
    scoring='neg_root_mean_squared_error',
    cv=5,                     # 5-fold cross-validation
    verbose=1,
    n_jobs=-1
)

random_search.fit(X_price_train, y_price_train)
```

```python
best_xgb_reg = random_search.best_estimator_

y_pred_best = best_xgb_reg.predict(X_price_test)


mse_best = mean_squared_error(y_price_test, y_pred_best)
r2_best = r2_score(y_test_reg, y_pred_best)

print(f"Tuned XGBRegressor RMSE: {mse_best ** 0.5:.2f}")
print(f"Tuned XGBRegressor RÂ²: {r2_best:.2f}")
```

```python
joblib.dump(best_xgb_reg, 'xgb_price_predictor.pkl')
```
